```python
import os
import cv2
import numpy as np 
import pandas as pd
```

```python
imagefolder='Dataset\images'
images=[]
labelslist=[]
size=(280,280)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def resize_and_save_images(image_paths, output_folder):
    for image_path in image_paths:
        img = cv2.imread(os.path.join(imagefolder, image_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        if img is not None:
            img_resized = cv2.resize(img, size)
            

            # Save resized image to the output folder
            output_path = os.path.join(output_folder, image_path)
            print(output_path)
            cv2.imwrite(output_path, img_resized)
        else:
            print(f"Failed to read or resize image: {image_path}")
```

```python
output_folder_healthy = 'Dataset/preprocessed_test'
os.mkdir('Dataset/preprocessed_test')
```

```python
test = os.listdir(imagefolder)
# print(test)
resize_and_save_images(test, output_folder_healthy)
```

```python
test_array = []
# print(output_folder_healthy)
test_images = os.listdir(output_folder_healthy)
c1, c2 = 0, 0
for image_path in test_images:
    img = cv2.imread(os.path.join(output_folder_healthy, image_path))
    test_array.append(img)
print(c1, c2)
# print(test_array)
test_arr = np.array(test_array)
```
